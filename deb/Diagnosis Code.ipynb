```python
def diag_classification(clm):
    if type(clm)==str:
        if (clm[0]=='E'):
            clm=clm[1:]
        elif (clm[0]=='V'):
            clm=clm[1:]
        elif len(clm)==5:
            clm = clm[:4]
    if int(float(clm))<1400:
        return 'infectious'
    elif (int(float(clm))>1400 and int(float(clm))<2400):
        return 'oncology'
    elif (int(float(clm))>2400 and int(float(clm))<2800):
        return 'endocrinology'
    elif (int(float(clm))>2800 and int(float(clm))<2900):
        return 'hematology'
    elif (int(float(clm))>2900 and int(float(clm))<3200):
        return 'psychiatry'
    elif (int(float(clm))>3200 and int(float(clm))<3900):
        return 'neurology'
    elif (int(float(clm))>3900 and int(float(clm))<4600):
        return 'cardiology'
    elif (int(float(clm))>4600 and int(float(clm))<5200):
        return 'pulmonology'
    elif (int(float(clm))>5200 and int(float(clm))<5800):
        return 'gastroenterology'
    elif (int(float(clm))>5800 and int(float(clm))<6300):
        return 'urology'
    elif (int(float(clm))>6300 and int(float(clm))<6800):
        return 'ob-gyn'
    elif (int(float(clm))>6800 and int(float(clm))<7100):
        return 'dermatology'
    elif (int(float(clm))>7100 and int(float(clm))<7400):
        return 'orthopedics'
    elif (int(float(clm))>7400 and int(float(clm))<7600):
        return 'congenital'
    elif (int(float(clm))>7600 and int(float(clm))<7800):
        return 'neonatology'
    elif (int(float(clm))>7800 and int(float(clm))<8000):
        return 'general'
    elif (int(float(clm))>8000 and int(float(clm))<9999):
        return 'emergency'
```

```python
df["diagnosis"] = df["DiagCode"].apply(lambda x: diag_classification(x))
```
